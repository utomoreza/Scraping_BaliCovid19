# Scrape using BeautifulSoup only

```python
# import all required libraries
import requests
import html5lib
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
```

```python
# enter URL address
url = 'https://covid19.badungkab.go.id/pemantauan-covid'
```

```python
# create BeautifulSoup object - start scraping
req_badung = requests.get(url)
soupBadung = BeautifulSoup(req_badung.content, 'html5lib')
```

```python
# get data from the table on webpage which is stored in 'table' tag with class attribute 'table table-bordered'
tables = soupBadung.findAll('table', attrs={'class':'table table-bordered'})
```

```python
# get kecamatan names
tbody = soupBadung.find('tbody', attrs={'class':None})

kec = []
for tr in tbody.findAll('tr', attrs={'class':'showModal'}):
    temp = tr.find('td', attrs={'class':None}).text
    temp = re.findall(r'[A-Z]+\s*[A-Z]*(?=\n)', temp)[0]
    kec.append(temp)
```

```python
listDesa = []
listKecamatan = []
listODPjumlah = []
listODPmasihPantau = []
listODPkeluarSelesai = []
listPDPjumlahKasus = []
listNegatif = []
listPositif = []
listBelumKeluar = []
listMasihRawat = []
listSembuh = []
listPosKasus = []
listPosRawat = []
listPosSembuh = []
listPosMeninggal = []
```

```python
# execute looping through each table
for tab_i in np.arange(1, len(tables)):
    # get the ith table from tables
    table = tables[tab_i]
    
    # get village name 'DESA' located in 'td' tag without any class attribute
    findDesa = table.findAll('td', attrs={'class':None})
    Desa = [] # create an empty list which will store every single village name of each district
    Kecamatan = []
    
    # execute looping for each district which will get all village names per district
    # village data start from 0th index along the way with 10 multiplication until the the length of findDesa minus 18
    for idx, i in enumerate(np.arange(0,len(findDesa)-18,10)):
        Desa.append(findDesa[i].text)
        Kecamatan.append(kec[tab_i-1]) # add kecamatan name for each desa
        
    # get 'ODP-JUMLAH KASUS (KOMULATIF)' data located in 'th' tag without any class attribute
    findODPjumlah = table.findAll('th', attrs={'class':None})
    ODPjumlah = [] 

    # execute looping for each district which will get all 'ODP-JUMLAH KASUS (KOMULATIF)' of each district
    # data of 'ODP-JUMLAH KASUS (KOMULATIF)' start from 2nd index along the way with 31 multiplication until the the length of findODPjumlah minus 30
    for i in np.arange(2,len(findODPjumlah)-30,31):
        ODPjumlah.append(int(findODPjumlah[i].text))
    
    # get 'ODP-MASIH PEMANTAUAN' data located in 'th' tag without any class attribute
    findODPmasihPantau = table.findAll('th', attrs={'class':None})
    ODPmasihPantau = [] 
    
    # execute looping for each district which will get all 'ODP-MASIH PEMANTAUAN' of each district
    # data of 'ODP-MASIH PEMANTAUAN' start from 5th index along the way with 31 multiplication until the the length of findODPmasihPantau minus 30
    for i in np.arange(5,len(findODPmasihPantau)-30,31):
        ODPmasihPantau.append(int(findODPmasihPantau[i].text))

    # get 'ODP-KELUAR/SELESAI PEMANTAUAN' data located in 'th' tag without any class attribute
    findODPkeluarSelesai = table.findAll('th', attrs={'class':None})
    ODPkeluarSelesai = [] 
    
    # execute looping for each district which will get all 'ODP-KELUAR/SELESAI PEMANTAUAN' of each district
    # data of 'ODP-KELUAR/SELESAI PEMANTAUAN' start from 8th index along the way with 31 multiplication until the the length of findODPkeluarSelesai minus 30
    for i in np.arange(8,len(findODPkeluarSelesai)-30,31):
        ODPkeluarSelesai.append(int(findODPkeluarSelesai[i].text))

    # get 'PDP-JUMLAH KASUS (KOMULATIF)' data located in 'th' tag without any class attribute
    findPDPjumlahKasus = table.findAll('th', attrs={'class':None})
    PDPjumlahKasus = [] 
    
    # execute looping for each district which will get all 'PDP-JUMLAH KASUS (KOMULATIF)' of each district
    # data of 'PDP-JUMLAH KASUS (KOMULATIF)' start from 11th index along the way with 31 multiplication until the the length of findPDPjumlahKasus minus 30
    for i in np.arange(11,len(findPDPjumlahKasus)-30,31):
        PDPjumlahKasus.append(int(findPDPjumlahKasus[i].text))

    # get 'NEGATIF' data located in 'td' tag without any class attribute
    findNegatif = table.findAll('td', attrs={'class': None})
    Negatif = []
    
    # execute looping for each district which will get all 'NEGATIF' of each district
    # data of 'NEGATIF' start from 3rd index along the way with 10 multiplication until the the length of findNegatif minus 15
    for i in np.arange(3,len(findNegatif)-15,10):
        Negatif.append(int(findNegatif[i].text))

    # get 'POSITIF' data located in 'td' tag without any class attribute
    findPositif = table.findAll('td', attrs={'class': None})
    Positif = []
    
    # execute looping for each district which will get all 'POSITIF' of each district
    # data of 'POSITIF' start from 6th index along the way with 10 multiplication until the the length of findPositif minus 12
    for i in np.arange(6,len(findPositif)-12,10):
        Positif.append(int(findPositif[i].text))

    # get 'BELUM KELUAR HASIL' data located in 'td' tag without any class attribute
    findBelumKeluar = table.findAll('td', attrs={'class': None})
    BelumKeluar = [] 
    # execute looping for each district which will get all 'BELUM KELUAR HASIL' of each district
    # data of 'BELUM KELUAR HASIL' start from 9th index along the way with 10 multiplication until the the length of findBelumKeluar minus 9
    for i in np.arange(9,len(findBelumKeluar)-9,10):
        BelumKeluar.append(int(findBelumKeluar[i].text))
    
    # get 'MASIH PERAWATAN' data located in 'th' tag without any class attribute
    findMasihRawat = table.findAll('th', attrs={'class':None})
    MasihRawat = [] 
    
    # execute looping for each district which will get all 'MASIH PERAWATAN' of each district
    # data of 'MASIH PERAWATAN' start from 15th index along the way with 31 multiplication until the the length of findMasihRawat minus 30
    for i in np.arange(15,len(findMasihRawat)-30,31):
        MasihRawat.append(int(findMasihRawat[i].text))
    
    # get 'SEMBUH' data located in 'th' tag without any class attribute
    findSembuh = table.findAll('th', attrs={'class':None})
    Sembuh = [] 
    
    # execute looping for each district which will get all 'SEMBUH' of each district
    # data of 'SEMBUH' start from 18th index along the way with 31 multiplication until the the length of findSembuh minus 30
    for i in np.arange(18,len(findSembuh)-30,31):
        Sembuh.append(int(findSembuh[i].text))

    # get 'POSITIF-KASUS' data located in 'th' tag without any class attribute
    findPosKasus = table.findAll('th', attrs={'class':None})
    PosKasus = []
        
    # execute looping for each district which will get all 'POSITIF-KASUS' of each district
    # data of 'POSITIF-KASUS' start from 21th index along the way with 31 multiplication until the the length of findPosKasus minus 30
    for i in np.arange(21,len(findPosKasus)-30,31):
        PosKasus.append(int(findPosKasus[i].text))
        
    # get 'POSITIF-DIRAWAT' data located in 'th' tag without any class attribute
    findPosRawat = table.findAll('th', attrs={'class':None})
    PosRawat = []
        
    # execute looping for each district which will get all 'POSITIF-DIRAWAT' of each district
    # data of 'POSITIF-DIRAWAT' start from 24th index along the way with 31 multiplication until the the length of findPosRawat minus 30
    for i in np.arange(24,len(findPosRawat)-30,31):
        PosRawat.append(int(findPosRawat[i].text))
        
    # get 'POSITIF-SEMBUH' data located in 'th' tag without any class attribute
    findPosSembuh = table.findAll('th', attrs={'class':None})
    PosSembuh = []
        
    # execute looping for each district which will get all 'POSITIF-SEMBUH' of each district
    # data of 'POSITIF-SEMBUH' start from 27th index along the way with 31 multiplication until the the length of findPosSembuh minus 30
    for i in np.arange(27,len(findPosSembuh)-30,31):
        PosSembuh.append(int(findPosSembuh[i].text))
        
    # get 'POSITIF-MENINGGAL' data located in 'th' tag without any class attribute
    findPosMeninggal = table.findAll('th', attrs={'class':None})
    PosMeninggal = []
        
    # execute looping for each district which will get all 'POSITIF-MENINGGAL' of each district
    # data of 'POSITIF-MENINGGAL' start from 30th index along the way with 31 multiplication until the the length of findPosMeninggal minus 30
    for i in np.arange(30,len(findPosMeninggal)-30,31):
        PosMeninggal.append(int(findPosMeninggal[i].text))
    
    # compile all
    listDesa.append(Desa)
    listKecamatan.append(Kecamatan)
    listODPjumlah.append(ODPjumlah)
    listODPmasihPantau.append(ODPmasihPantau)
    listODPkeluarSelesai.append(ODPkeluarSelesai)
    listPDPjumlahKasus.append(PDPjumlahKasus)
    listNegatif.append(Negatif)
    listPositif.append(Positif)
    listBelumKeluar.append(BelumKeluar)
    listMasihRawat.append(MasihRawat)
    listSembuh.append(Sembuh)
    listPosKasus.append(PosKasus)
    listPosRawat.append(PosRawat)
    listPosSembuh.append(PosSembuh)
    listPosMeninggal.append(PosMeninggal)
```

```python
# since the results above are still in 2D lists, we need to flatten them into 1D lists

desaList = []
kecamatanList = []
ODPjumlahList = []
ODPmasihPantauList = []
ODPkeluarSelesaiList = []
PDPjumlahKasusList = []
NegatifList = []
PositifList = []
BelumKeluarList = []
MasihRawatList = []
SembuhList = []
PosKasusList = []
PosRawatList = []
PosSembuhList = []
PosMeninggalList = []

# then execute looping for each 2D list in order to get 1D version list
for i, i_fill in enumerate(listDesa):
    for j, j_fill in enumerate(listDesa[i]):
        desaList.append(listDesa[i][j])
        
for i, i_fill in enumerate(listKecamatan):
    for j, j_fill in enumerate(listKecamatan[i]):
        kecamatanList.append(listKecamatan[i][j])
        
for i, i_fill in enumerate(listODPjumlah):
    for j, j_fill in enumerate(listODPjumlah[i]):
        ODPjumlahList.append(listODPjumlah[i][j])
        
for i, i_fill in enumerate(listODPmasihPantau):
    for j, j_fill in enumerate(listODPmasihPantau[i]):
        ODPmasihPantauList.append(listODPmasihPantau[i][j])
        
for i, i_fill in enumerate(listODPkeluarSelesai):
    for j, j_fill in enumerate(listODPkeluarSelesai[i]):
        ODPkeluarSelesaiList.append(listODPkeluarSelesai[i][j])
        
for i, i_fill in enumerate(listPDPjumlahKasus):
    for j, j_fill in enumerate(listPDPjumlahKasus[i]):
        PDPjumlahKasusList.append(listPDPjumlahKasus[i][j])
        
for i, i_fill in enumerate(listNegatif):
    for j, j_fill in enumerate(listNegatif[i]):
        NegatifList.append(listNegatif[i][j])
        
for i, i_fill in enumerate(listPositif):
    for j, j_fill in enumerate(listPositif[i]):
        PositifList.append(listPositif[i][j])
        
for i, i_fill in enumerate(listBelumKeluar):
    for j, j_fill in enumerate(listBelumKeluar[i]):
        BelumKeluarList.append(listBelumKeluar[i][j])
        
for i, i_fill in enumerate(listMasihRawat):
    for j, j_fill in enumerate(listMasihRawat[i]):
        MasihRawatList.append(listMasihRawat[i][j])
        
for i, i_fill in enumerate(listSembuh):
    for j, j_fill in enumerate(listSembuh[i]):
        SembuhList.append(listSembuh[i][j])
        
for i, i_fill in enumerate(listPosKasus):
    for j, j_fill in enumerate(listPosKasus[i]):
        PosKasusList.append(listPosKasus[i][j])
        
for i, i_fill in enumerate(listPosRawat):
    for j, j_fill in enumerate(listPosRawat[i]):
        PosRawatList.append(listPosRawat[i][j])
        
for i, i_fill in enumerate(listPosSembuh):
    for j, j_fill in enumerate(listPosSembuh[i]):
        PosSembuhList.append(listPosSembuh[i][j])
        
for i, i_fill in enumerate(listPosMeninggal):
    for j, j_fill in enumerate(listPosMeninggal[i]):
        PosMeninggalList.append(listPosMeninggal[i][j])
```

```python
# after all 1D lists generated, now combine all into a final list then set as a dataframe (and transpose)
dataBadung = [kecamatanList, desaList, ODPjumlahList, ODPmasihPantauList, ODPkeluarSelesaiList, PDPjumlahKasusList, 
              NegatifList, PositifList, BelumKeluarList, MasihRawatList, SembuhList, PosKasusList, PosRawatList, PosSembuhList, PosMeninggalList]

df = pd.DataFrame(dataBadung).transpose()
```

```python
# define the headers
headersBadung = ['TANGGAL-UPDATE', 'KABUPATEN', 'KECAMATAN', 'DESA', 'ODP-JUMLAH KASUS KUMULATIF', 'ODP-MASIH PEMANTAUAN', 
                 'ODP-KELUAR/SELESAI PEMANTAUAN', 'PDP-JUMLAH KASUS KUMULATIF', 'PDP-NEGATIF', 
                 'PDP-POSITIF', 'PDP-BELUM KELUAR HASIL', 'PDP-MASIH PERAWATAN', 'PDP-SEMBUH',
                 'POSITIF-KASUS','POSITIF-DIRAWAT','POSITIF-SEMBUH','POSITIF-MENINGGAL']
```

```python
# make a list filled with 'Badung' regency
Kabupaten = ['BADUNG' for i in range(df.shape[0])]
```

```python
# get date information obtained from the webpage
# it's located in the 2nd index for 'h5' tag with attribute class 'text-center'
DateRaw = soupBadung.findAll('h5', class_='text-center')[1].text
DateRaw = DateRaw[4:] # clean the date
```

```python
# create a date list
Date = [DateRaw for i in range(df.shape[0])]
```

```python
# insert columns of 'TANGGAL-UPDATE', 'KABUPATEN', and 'KECAMATAN' to the df
df.insert(0, column='TANGGAL-UPDATE', value=Date)
df.insert(1, column='KABUPATEN', value=Kabupaten)
# df.insert(2, column='KECAMATAN', value=Kecamatan)
```

```python
# set headers
df.columns = headersBadung
```
